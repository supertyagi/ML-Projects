```python
from sklearn.model_selection import train_test_split
```

```python
X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)
```

```python
from sklearn.preprocessing import Imputer
```

```python
missing_values = Imputer(missing_values = 0, strategy = 'mean', axis = 0)

X_train = missing_values.fit_transform(X_train)
X_test = missing_values.fit_transform(X_test)
```

```python
X_train[0]
```

```python
from sklearn.ensemble import RandomForestClassifier

classifier = RandomForestClassifier(random_state=10)

classifier.fit(X_train, y_train.ravel())
```

```python
predict_train_data = classifier.predict(X_test)

from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

print('accuracy is ', accuracy_score(y_test, predict_train_data))
```

```python
def timer(start_time=None):
    if not start_time:
        start_time = datetime.now()
        return start_time
    elif start_time:
        thour, temp_sec = divmod((datetime.now() - start_time).total_seconds(), 3600)
        tmin, tsec = divmod(temp_sec, 60)
        print('\n Time taken: %i hours %i minutes and %s seconds.' % (thour, tmin, round(tsec, 2)))
```

```python
from datetime import datetime
# Here we go
start_time = timer(None) # timing starts from this point for "start_time" variable
random_search.fit(X_train,y_train.ravel())
timer(start_time) # timing ends here for "start_time" variable
```
