### Find Missing Values

```python
columns_with_nan = [column for column in df.columns if df[column].isnull().sum() >= 1]
```

```python
total_columns_with_nan = 0
for column in columns_with_nan:
    if df[column].isnull().sum()>=1:
        missing_percentage = df[column].isnull().mean()
        print(column,'has',np.round(missing_percentage, 4),'% missing values')
        total_columns_with_nan+=1
print('Total columns with nan value : ',total_columns_with_nan)
```

```python
numerical_columns = [column for column in df.columns if df[column].dtypes != 'O']
print(numerical_columns)
print(len(numerical_columns))
```

```python
year_columns = [column for column in df.columns if 'Yr' in column or 'Year' in column]
print(year_columns)
```

```python
for year in year_columns:
    if year != 'YrSold':
        df[year] = df['YrSold']-df[year]
```

```python
df.head()
```

```python
## Numerical variables are usually of 2 type
## 1. Continous variable and Discrete Variables

discrete_columns = [column for column in numerical_columns if len(df[column].unique())<25 and column not in year_columns+['Id']]
print("Discrete columns Count: {}".format(len(discrete_columns)))
```

```python
discrete_columns
```

```python
## Lets Find the realtionship between them and Sale PRice

for column in discrete_columns:
    data=df.copy()
    data.groupby(column)['SalePrice'].median().plot.bar()
    plt.xlabel(column)
    plt.ylabel('SalePrice')
    plt.title(column)
    plt.show()
```

```python
continuous_columns = [column for column in numerical_columns if column not in discrete_columns+year_columns+['Id']]
print("Continuous column Count {}".format(len(continuous_columns)))
```

```python
len(numerical_columns)
```

```python
for column in continuous_columns:
    data=df.copy()
    data[column].hist(bins=25)
    plt.xlabel(column)
    plt.ylabel("Count")
    plt.title(column)
    plt.show()
```

```python
## We will be using logarithmic transformation


for column in continuous_columns:
    data=df.copy()
    if 0 in data[column].unique():
        pass
    else:
        data[column]=np.log(data[column])
        data['SalePrice']=np.log(data['SalePrice'])
        plt.scatter(data[column],data['SalePrice'])
        plt.xlabel(column)
        plt.ylabel('SalesPrice')
        plt.title(column)
        plt.show()
```

```python
#Checking the Outliers

for column in continuous_columns:
    data=df.copy()
    if 0 in data[column].unique():
        pass
    else:
        data[column]=np.log(data[column])
        data.boxplot(column=column)
        plt.ylabel(column)
        plt.title(column)
        plt.show()
```
